```python
import os
import cv2
import numpy as np
from sklearn.model_selection import train_test_split

# Function to load images from subfolders inside the dataset
def load_images_from_subfolders(root_path, img_size):
    images = []
    labels = []
    for subfolder in os.listdir(root_path):
        subfolder_path = os.path.join(root_path, subfolder)
        if os.path.isdir(subfolder_path):
            for file_name in os.listdir(subfolder_path):
                file_path = os.path.join(subfolder_path, file_name)
                if file_name.endswith('.jpg') or file_name.endswith('.png'):
                    # Load and resize image
                    img = cv2.imread(file_path, cv2.IMREAD_GRAYSCALE)
                    img = cv2.resize(img, (img_size, img_size))
                    img = img / 255.0  # Normalize pixel values

                    images.append(img)
                    labels.append(subfolder)  # Store the label based on the folder name
    return np.array(images), np.array(labels)

# Load images from normal_dataset folder
image_size = 128  # Resize images to 128x128
dataset_path = "alzheimers_dataset/dataset/normal_dataset"
images, labels = load_images_from_subfolders(dataset_path, image_size)

# Add channel dimension for the images
images = np.expand_dims(images, axis=-1)

# Split into training and test sets
train_images, test_images, train_labels, test_labels = train_test_split(images, labels, test_size=0.2, random_state=42)

# Verify the shapes
print(f"Training Images: {train_images.shape}")
print(f"Testing Images: {test_images.shape}")
```

```python
import os
import numpy as np
import cv2
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.layers import Input, Conv2D, BatchNormalization, ReLU, Add
from tensorflow.keras.models import Model

# Parameters
image_size = 128  # Resize images to 128x128
batch_size = 32
epochs = 50
dataset_path = '/content/alzheimers_dataset/dataset/normal_dataset/NonDemented'  # Your dataset path

# Load and preprocess dataset
def load_images(dataset_path, img_size):
    images = []
    for file_name in os.listdir(dataset_path):
        file_path = os.path.join(dataset_path, file_name)
        if file_name.endswith('.png') or file_name.endswith('.jpg'):
            img = cv2.imread(file_path, cv2.IMREAD_GRAYSCALE)
            img = cv2.resize(img, (img_size, img_size))
            img = img / 255.0  # Normalize
            images.append(img)
    return np.array(images)

# Load dataset
images = load_images(dataset_path, image_size)
images = np.expand_dims(images, axis=-1)  # Add channel dimension
print(f"Loaded images shape: {images.shape}")

# Split dataset into training and testing sets
train_images, test_images = train_test_split(images, test_size=0.2, random_state=42)
print(f"Training images shape: {train_images.shape}")
print(f"Testing images shape: {test_images.shape}")

# Build the CNN model
def build_model(input_shape):
    inputs = Input(shape=input_shape)

    # Convolutional Block 1
    x = Conv2D(64, (3, 3), padding='same')(inputs)
    x = ReLU()(x)

    # Residual Blocks
    for _ in range(5):
        res = Conv2D(64, (3, 3), padding='same')(x)
        res = BatchNormalization()(res)
        res = ReLU()(res)
        res = Conv2D(64, (3, 3), padding='same')(res)
        res = BatchNormalization()(res)
        x = Add()([x, res])

    # Convolutional Block 2
    x = Conv2D(1, (3, 3), padding='same')(x)
    outputs = ReLU()(x)

    return Model(inputs, outputs)

# Compile and train the model
model = build_model((image_size, image_size, 1))
model.compile(optimizer='adam', loss='mse', metrics=['mae'])

# Train the model
model.fit(
    x=train_images,
    y=train_images,  # Autoencoder: input = output
    validation_data=(test_images, test_images),
    batch_size=batch_size,
    epochs=epochs
)

# Save the model
model.save("medical_image_enhancer.h5")

# Enhance test images
enhanced_images = model.predict(test_images)

# Save enhanced images
output_path = "enhanced_images"
os.makedirs(output_path, exist_ok=True)
for i, img in enumerate(enhanced_images):
    img = (img.squeeze() * 255).astype(np.uint8)
    cv2.imwrite(os.path.join(output_path, f"enhanced_{i}.png"), img)

print(f"Enhanced images saved in {output_path}")
```
